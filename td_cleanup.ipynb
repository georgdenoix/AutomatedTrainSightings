```python
import os, shutil
from datetime import datetime
```

```python
#root_folder = 'C:\\Users\\georg\\Pictures\\TrainDetector'

root_folder = 'c:\\users\\georg\\Documents\\Rail\\TrainDetector'
```

```python
os.listdir(root_folder)
```

```python
no_train_folder = os.path.join(root_folder, 'results', 'media', 'no_train')
inbox_folder = os.path.join(root_folder, 'inbox')
```

```python
cam_list = []
for cam in os.listdir(no_train_folder):
    cam_list.append(cam)
```

## Clean up "no_train" folders

```python
for cam in cam_list:
    cam_folder = os.path.join(no_train_folder, cam)
    print(cam_folder)
    try:
        for datestr in os.listdir(cam_folder):
            cam_date = datetime.strptime(datestr, '%Y-%m-%d')
            age = datetime.now() - cam_date
            #if (len(cam) == 4 and age.days > 7) or (len(cam) == 3 and age.days > 14):
            if age.days > 7:
                folder_to_delete = os.path.join(cam_folder, datestr)
                print('deleting:', folder_to_delete)
                shutil.rmtree(folder_to_delete)
    except:
        pass
```

## Clean up "inbox" folders

```python
cam_list = []
for cam in os.listdir(inbox_folder):
    cam_list.append(cam)
```
