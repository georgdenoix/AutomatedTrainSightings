## Simple filters

```python
start_date = pd.to_datetime('2023-05-01')
end_date = pd.to_datetime('2023-05-02')
```

```python
# filter by date range
filtered_data = df.loc[(pd.to_datetime(df['date']) >= start_date) & (pd.to_datetime(df['date']) <= end_date) & (df['shows_train'] == 1)]
```

```python
# filter by camera number
filtered_data = df.loc[(df['shows_train'] == 1) & (df['camera'] == 268)]
```

```python
#reconstruct path to file
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

pid = 118

root_folder = 'C:\\Users\\georg\\Pictures\\TrainDetector'

try:
    if len(filtered_data.iloc[pid]['file_path']) > 0:
        file_path = filtered_data.iloc[pid]['file_path']
except:
    file_path = os.path.join(root_folder, 'results', 'media', 'train', str(filtered_data.iloc[pid]['camera']), filtered_data.iloc[pid]['date'])

file_name = filtered_data.iloc[pid]['file_name']

print(file_name)

# if file name points to 
if file_name[-4:] == '.mp4':
    file_name = file_name[:-4] + '_0.jpg'
elif (file_name[-5:] == '.jpeg') or (file_name[-4:] == '.jpg'):
    pass
elif filtered_data.iloc[pid]['camera'] >= 3000:
    file_name += '_0.jpg'
else:
    file_name += '.jpeg'
```

```python
image_path = os.path.join(file_path, file_name)

# Load the image
img = mpimg.imread(image_path)

# Create a figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(img)
ax.axis('off')  # Optional: Hide the axis

# Add a title and description
title = filtered_data.iloc[pid]['caption']
description = f'cam {filtered_data.iloc[pid]["camera"]} - {filtered_data.iloc[pid]["date"]} at {filtered_data.iloc[pid]["time"]}'
ax.set_title(title)
ax.text(0.5, -0.1, description, transform=ax.transAxes,
        ha='center', fontsize=10, color='gray')

# Show the plot
plt.show()
```
